```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier  # Import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the training dataset
train_data = pd.read_csv('fraudTrain.csv')
train_data
```

```python
# Explore the training dataset
print(train_data.info())
```

```python
print(train_data.describe())
```

```python
train_data.dtypes
```

```python
# Check for missing values
print(train_data.isnull().sum())
```

```python
train_data.columns
```

```python
# Convert date columns to datetime
train_data["trans_date_trans_time"] = pd.to_datetime(train_data["trans_date_trans_time"])
train_data["dob"] = pd.to_datetime(train_data["dob"])
```

```python
# Extract relevant information from datetime columns
train_data["trans_year"] = train_data["trans_date_trans_time"].dt.year
train_data["trans_month"] = train_data["trans_date_trans_time"].dt.month
train_data["trans_day"] = train_data["trans_date_trans_time"].dt.day
train_data["trans_hour"] = train_data["trans_date_trans_time"].dt.hour
train_data["dob_year"] = train_data["dob"].dt.year
train_data["dob_month"] = train_data["dob"].dt.month
train_data["dob_day"] = train_data["dob"].dt.day


# Display the modified DataFrame
print(train_data)
```

```python
train_data.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],inplace=True)
train_data
```

```python
# Drop rows with missing values
train_data.dropna(ignore_index=True, inplace=True)
train_data
```

```python
# Label encode categorical variables
encoder = LabelEncoder()
train_data["merchant"] = encoder.fit_transform(train_data["merchant"])
train_data["category"] = encoder.fit_transform(train_data["category"])
train_data["gender"] = encoder.fit_transform(train_data["gender"])
train_data["job"] = encoder.fit_transform(train_data["job"])
```

```python
# Display is_fraud counts
exit_counts = train_data["is_fraud"].value_counts()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pie(exit_counts, labels=["No", "YES"], autopct="%0.0f%%")
plt.title("fraud Counts")
plt.tight_layout()
plt.show()
```

```python
# Prepare features (X) and target variable (Y)
X = train_data.drop(columns=["is_fraud"])
Y = train_data["is_fraud"]
```

```python
# Create Random Forest model
model = RandomForestClassifier()  # You can adjust hyperparameters if needed
model.fit(X, Y)
```

```python
# Check the model's training accuracy
train_accuracy = accuracy_score(Y, model.predict(X))
print("Training Accuracy:", train_accuracy)
```

```python
# Correlation Matrix
correlation_matrix = train_data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
